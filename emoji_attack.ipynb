```python

from sentence_transformers import SentenceTransformer, util
import numpy as np
import pandas as pd
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [
    "I love pizza",
    " I  love  pizza ",
    "I lo ve pizza",
    "I love piz za",
    "I love p izza",
    "I love pi zza",
    "I love pizza üçï",
    "I love pizza üçïüçï",
    "üçïI üçïlove üçïpizza ",
    "I love pizza üòç",
    "I love pizza üò¢",
    "I love pizza üòÇ",
    "I love pizüçïza ",
    "I love piüçïzza üçï",
    "I love püçïizza üçï"
]
```

```python
embeddings = model.encode(sentences, convert_to_tensor=True)
```

```python
emb_np = embeddings.cpu().numpy()
emb_np
```

```python
cosine_sim = util.cos_sim(embeddings, embeddings)
```

```python
df = pd.DataFrame(cosine_sim.cpu().numpy(),
                  index=sentences,
                  columns=sentences)
display(df)
```

```python
base_vector = emb_np[0]  # "I love pizza"
shifts = [np.linalg.norm(vec - base_vector) for vec in emb_np]
```

```python
shift_df = pd.DataFrame({"Sentence": sentences, "Vector Shift (L2 Distance)": shifts})
print("\nEmbedding Shifts Compared to Base Sentence:")
display(shift_df)
```
